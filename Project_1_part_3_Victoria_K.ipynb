```python
#importing the dataset
Sales_prediction='/content/drive/MyDrive/Colab data uploads/sales_predictions_2023.csv'
```

```python
import pandas as pd
df=pd.read_csv(Sales_prediction)
```

```python
df.head()
```

Histograms to view the distributions of numerical features in your dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

Boxplots to view statistical summaries of numerical features in your dataset.

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    plt.figure(figsize=(10, 8))
    df[numerical_columns].boxplot()
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()
```

Countplots to view the frequency of each class of categorial features in your dataset.

```python
import seaborn as sns
```

```python
 #Select categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Plot countplots for each categorical feature
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column, palette='pastel')
    plt.title(f'Countplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

Heatmap to view the correlation between features.

```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.show()
```
